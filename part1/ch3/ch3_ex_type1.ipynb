### ğŸŠ Section12. ê²°ì¸¡ì¹˜ ì²˜ë¦¬, ìµœë¹ˆê°’, ë°ì´í„° ê°œìˆ˜

#### â˜‘ï¸ <span style='background-color: #fff5b1'>df.dropna()</span>

```python
# SOL 1) ê²°ì¸¡ì¹˜ê°€ ê°€ì¥ ë§ì€ ì»¬ëŸ¼ ìˆ˜ë™ìœ¼ë¡œ í™•ì¸

import pandas as pd
df = pd.read_csv("type1_data1.csv")

# ê²°ì¸¡ì¹˜ê°€ ê°€ì¥ ë§ì€ ë‘ ì»¬ëŸ¼ ì°¾ê¸° => f1, f3
df.isnull().sum()

# ì²«ë²ˆì§¸ë¡œ ê²°ì¸¡ì¹˜ê°€ ë§ì€ ì»¬ëŸ¼ì—ì„œ ê²°ì¸¡ì¹˜ê°€ ìˆëŠ” í–‰ ì œê±°
df = df.dropna(subset=['f1'])

# ë‘ë²ˆì§¸ë¡œ ê²°ì¸¡ì¹˜ê°€ ë§ì€ ì»¬ëŸ¼ì„ ìµœë¹ˆê°’ìœ¼ë¡œ ëŒ€ì²´
freq = df['f3'].mode()[0]
df['f3'] = df['f3'].fillna(freq)

# 'f3'ì»¬ëŸ¼ì˜ 'gold'ê°’ì„ ê°€ì§„ ë°ì´í„°ì˜ ìˆ˜ë¥¼ ì •ìˆ˜í˜•ìœ¼ë¡œ êµ¬í•˜ê¸°
print(len(df[df['f3']=='gold']))
print(sum(df['f3']=='gold'))
```

```python
# SOL 2) ê²°ì¸¡ì¹˜ê°€ ê°€ì¥ ë§ì€ ì»¬ëŸ¼ idxmax()ë¡œ êµ¬í•˜ê¸°

import pandas as pd
df = pd.read_csv("type1_data1.csv")

# ê²°ì¸¡ì¹˜ê°€ ê°€ì¥ ë§ì€ ì»¬ëŸ¼
col = df.isnull().sum().idxmax()

# ê²°ì¸¡ì¹˜ê°€ ê°€ì¥ ë§ì€ ì»¬ëŸ¼ ì œê±°
df = df.dropna(subset=[col])

# ë‘ë²ˆì§¸ë¡œ ê°€ì¥ ê²°ì¸¡ì¹˜ê°€ ë§ì€ ì»¬ëŸ¼ ìµœë¹ˆê°’ ëŒ€ì²´
freq = df['f3'].mode()[0]
df['f3'] = df['f3'].fillna(freq)

# f3ì»¬ëŸ¼ì´ goldì¸ ë°ì´í„°ì˜ ìˆ˜
print(len(df[df['f3'] == 'gold']))
```

### ğŸ¥¦ Section16. ì´ìƒì¹˜, IQR

### ğŸŠ Section25. ì‹œê°„ ê°„ì˜ ì°¨ì´ ê³„ì‚°(ì¼)

#### â˜‘ï¸ <span style='background-color: #fff5b1'>.assign() / .transform() / .agg() / .apply()</span>

```python
# SOL 1)

import pandas as pd
df = pd.read_csv("delivery_time.csv")

# ê° ì‚¬ìš©ìë³„ë¡œ ì²« ì£¼ë¬¸ê³¼ ë§ˆì§€ë§‰ ì£¼ë¬¸ ì‚¬ì´ì˜ ì‹œê°„ ê°„ê²©ì„ ì¼ ë‹¨ìœ„ë¡œ ê³„ì‚°
df['ì£¼ë¬¸ì‹œê°„'] = pd.to_datetime(df['ì£¼ë¬¸ì‹œê°„'])
min_order_time = df.groupby('user')['ì£¼ë¬¸ì‹œê°„'].min()
max_order_time = df.groupby('user')['ì£¼ë¬¸ì‹œê°„'].max()
time_interval = (max_order_time - min_order_time).dt.days

# ì‹œê°„ì°¨ê°€ 0ì¼ì¸ ì‚¬ìš©ìë¥¼ ì œì™¸í•˜ê³ , ë‚˜ë¨¸ì§€ ì‚¬ìš©ìë“¤ì˜ í‰ê·  ì‹œê°„ ê°„ê²©(ì¼ ë‹¨ìœ„) ê³„ì‚°
cond1 = time_interval > 0
m = time_interval[cond1].mean()

# í‰ê·  ì‹œê°„ ê°„ê²©ë³´ë‹¤ ê¸´ ì‹œê°„ ê°„ê²©ì„ ê°€ì§„ ì‚¬ìš©ìì˜ ìˆ˜ë¥¼ ì •ìˆ˜ë¡œ êµ¬í•˜ê¸°
cond2 = time_interval > m
print(len(time_interval[cond2]))
```

```python
# SOL 2)

import pandas as pd

df = pd.read_csv("delivery_time.csv")
df['ì£¼ë¬¸ì‹œê°„'] = pd.to_datetime(df['ì£¼ë¬¸ì‹œê°„'])

# ì‚¬ìš©ìë³„ ìµœì´ˆ, ìµœì¢… ì£¼ë¬¸ì‹œê°„ì„ ì›ë³¸ dfì— ì¶”ê°€
df = df.assign(
    first_order=df.groupby('user')['ì£¼ë¬¸ì‹œê°„'].transform('min'),
    last_order=df.groupby('user')['ì£¼ë¬¸ì‹œê°„'].transform('max')
)

# ì‹œê°„ì°¨ ê³„ì‚° (ì¼ ë‹¨ìœ„)
df['interval_days'] = (df['last_order'] - df['first_order']).dt.days

# ì‚¬ìš©ìë³„ ì‹œê°„ì°¨ëŠ” ëª¨ë‘ ë™ì¼í•˜ë¯€ë¡œ drop_duplicatesë¡œ ì¤‘ë³µ ì œê±°
user_intervals = df[['user', 'interval_days']].drop_duplicates()

# ì‹œê°„ì°¨ê°€ 0ì¼ ì´ˆê³¼ì¸ ì‚¬ìš©ìë“¤ë§Œ ì¶”ë ¤ í‰ê·  ê³„ì‚°
valid_users = user_intervals[user_intervals['interval_days'] > 0]
avg_days = valid_users['interval_days'].mean()

# í‰ê· ë³´ë‹¤ í° ì‚¬ìš©ì ìˆ˜ ê³„ì‚°
count = (valid_users['interval_days'] > avg_days).sum()
print(int(count))
df
```

```python
import pandas as pd
df = pd.read_csv("region_sales.csv")

# í”¼ë²— í…Œì´ë¸” ìƒì„±: Region, Channelë³„ë¡œ Productë³„ Sales í•©ê³„ ê³„ì‚°
pivot_table = pd.pivot_table(df, index=['Region', 'Channel'], columns='Product',
                             values='Sales', aggfunc='sum')

# Aë¹„ìœ¨ ê³„ì‚°: ì œí’ˆ Aì˜ íŒë§¤ì•¡ Ã· ì´ë§¤ì¶œ
pivot_table['ì´ë§¤ì¶œ'] = pivot_table['A'] + pivot_table['B']
pivot_table['Aë¹„ìœ¨'] = pivot_table['A']/pivot_table['ì´ë§¤ì¶œ']

# ìµœëŒ€ Aë¹„ìœ¨ ê°’ ì¶œë ¥
print(round(pivot_table['Aë¹„ìœ¨'].max(), 2))
```

```python
import pandas as pd
df = pd.read_csv('monthly_sales.csv')

# meltë¡œ Wide(ê°€ë¡œ) â†’ Long(ì„¸ë¡œ) ë³€í™˜
melted_df = pd.melt(df, id_vars='Region', value_vars=['Jan', 'Feb', 'Mar'],
                    var_name='Month', value_name='Sales')

# Region, Monthë³„ ë§¤ì¶œ í•©ê³„ êµ¬í•˜ê¸°
df1 = melted_df.groupby(['Region', 'Month'])['Sales'].sum().reset_index()

# ë§¤ì¶œ í•©ê³„ê°€ 1400ì„ ì´ˆê³¼í•˜ëŠ” ê±´ìˆ˜
print(sum(df1['Sales']>1400))
```
