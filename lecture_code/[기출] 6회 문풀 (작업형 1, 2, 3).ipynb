3. 연도별로 총 범죄 건수(범죄유형의 총합)의 월평균 값을 구한 후 그 값이 가장 큰 연도를 찾아, 해당 연도의 총 범죄 건수의 월평균 값을 출력하시오. (반올림하여 정수로 출력)
- **총 범죄건수의 월평균 값을 구할 때 모든 월의 데이터가 있는 것이 아니므로 .sum()으로 합을 구한 후 12로 나눠줘야 함.**

```python
# My)

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_1/data6-1-3.csv")

df['연도'] = df['날짜'].str[:4].astype(int)
df['월'] = df['날짜'].str[-3:-1].astype(int)
df = df.drop(['날짜'], axis=1)
df['총범죄건수'] = df.iloc[:, :6].sum(axis=1)
# print(df)

g1 = df.groupby(['연도'])['총범죄건수'].sum()/12
top_year = g1.idxmax()

df1 = df[df['연도']==top_year]
# print(df1)
print(round(df1['총범죄건수'].mean()))
```

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_2/energy_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_2/energy_test.csv")

# 1. 데이터 유형 파악
# 결측값: X
# 범주형: ['Roof', 'Height', 'Orient']
# target: Heat_Load

# print(train.info())
# print(test.info())
# print(train.shape, test.shape) # (537, 10) (231, 9)
# print(train.select_dtypes('object').columns)

# 2. 데이터 전처리
# (1) X_full
X = train.drop(['Heat_Load'], axis=1)
y = train['Heat_Load']
X_full = pd.concat([X, test], axis=0)
# print(X.shape, test.shape, y.shape) # (537, 9) (231, 9) (537,)

# (2) 결측값 처리 => X

# (3) 수치형 변수 스케일링 => X

# (4) 범주형 변수 인코딩
X_full = pd.get_dummies(X_full)
# print(X_full)
# print(X_full.shape) # (768, 16)

# 3. 데이터 분리
X_train = X_full[:train.shape[0]]
X_test = X_full[train.shape[0]:]

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2)
# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape) # (429, 16) (108, 16) (429,) (108,)

# 4. 모델 학습 및 검증
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_val_pred_rfc = rfc.predict(X_val)
# print(y_val_pred_rfc)
# print(y_val_pred_rfc.shape) # (108,)

from sklearn.ensemble import HistGradientBoostingClassifier
hgbc = HistGradientBoostingClassifier()
hgbc.fit(X_train, y_train)
y_val_pred_hgbc = hgbc.predict(X_val)
# print(y_val_pred_hgbc)
# print(y_val_pred_hgbc.shape) # (108,)

# 5. 모델 평가
from sklearn.metrics import f1_score
rfc_score = f1_score(y_val, y_val_pred_rfc, average='macro')
hgbc_score = f1_score(y_val, y_val_pred_hgbc, average='macro')
print('rfc_score', rfc_score)
print('hgbc_score', hgbc_score)

# 6. 결과 저장
y_pred = rfc.predict(X_test)
# print(y_pred)
# print(y_pred.shape) # (231,)

result = pd.DataFrame({'pred':y_pred})
result.to_csv('result_62.csv', index=False)

check = pd.read_csv('result_62.csv')
print(check)
```

## 문제2. 다중 선형 회귀 문제

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_3/data6-3-2.csv")
df.head(2)
```

1. 다중 선형 회귀 모델을 구축하고, 독립변수 o3의 회귀계수를 구하시오.
- 독립변수: solar(태양 에너지), wind(바람의 세기), o3(오존 농도)
- 종속변수: temperature(온도)


```python
from statsmodels.formula.api import ols
formula = "temperature ~ solar + wind + o3"
model = ols(formula, data=df).fit()
print(model.summary())
# 0.0749
model.params['o3']
```

2. 데이터에서 'solar'와 'o3' 값을 고정한 상태에서, 'wind'의 세기가 증가함에 따라 'temperature'가 감소하는지를 검증하기 위해 다중 선형 회귀 분석을 수행하고, 이 때'wind'의 회귀 계수에 대한 p-value 값을 구하시오. (유의수준: 0.05)

```python
0.78
model.pvalues['wind']
```

3. solar:100, wind:5, o3:30일 때 예측값과 그에 대한 95% 신뢰구간을 구하시오.

```python
new_data = pd.DataFrame({
    'solar':[100],
    'wind':[5],
    'o3':[30]
})
pred = model.get_prediction(new_data)
print(pred.summary_frame(alpha=0.05))
```
