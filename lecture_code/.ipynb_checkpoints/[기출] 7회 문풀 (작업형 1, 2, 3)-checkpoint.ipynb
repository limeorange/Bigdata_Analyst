## ë¬¸ì œ3
IQRì„ ì´ìš©í•´ ì´ì‚°í™”íƒ„ì†Œ(CO2) ì´ìƒì¹˜ ìˆ˜ë¥¼ ì°¾ìœ¼ì‹œì˜¤

- Q1 (ì œ 1ì‚¬ë¶„ìœ„ìˆ˜): í•˜ìœ„ 25% ê°’
- Q3 (ì œ 3ì‚¬ë¶„ìœ„ìˆ˜): ìƒìœ„ 25% ê°’
- IQR (ì‚¬ë¶„ìœ„ìˆ˜ ë²”ìœ„): IQR = Q3 - Q1

- lower (í•˜í•œì„ ): Q1 - 1.5 * IQR, ì´ ê°’ë³´ë‹¤ ì‘ì€ ë°ì´í„°ëŠ” ì´ìƒì¹˜ë¡œ ê°„ì£¼
- upper (ìƒí•œì„ ): Q3 + 1.5 * IQR, ì´ ê°’ë³´ë‹¤ í° ë°ì´í„°ëŠ” ì´ìƒì¹˜ë¡œ ê°„ì£¼



```python
# My) 

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/7_1/air_quality.csv")

# print(df)
Q1 = df['CO2'].quantile(.25)
Q3 = df['CO2'].quantile(.75)
IQR = Q3 - Q1

lower = Q1-1.5*IQR
upper = Q3+1.5*IQR

cond1 = df['CO2'] < lower
cond2 = df['CO2'] > upper
outliers = df[cond1 | cond2]

# print(Q1, Q3, IQR)
print(len(outliers))
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/7_1/air_quality.csv")

# 1) IQR ê³„ì‚°
Q1 = df['CO2'].quantile(0.25)
Q3 = df['CO2'].quantile(0.75)
IQR = Q3 - Q1

# 2) ìƒí•œ ë° í•˜í•œ ê³„ì‚°
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR

# 3) ì´ìƒì¹˜ ì¡°ê±´
cond1 = df['CO2'] < lower
cond2 = df['CO2'] > upper
outliers = df[cond1 | cond2]

# 4) ì´ìƒì¹˜ ìˆ˜ ê³„ì‚°
print(len(outliers))
```

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/7_2/mart_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/7_2/mart_test.csv")

# 1. ë°ì´í„° ìœ í˜• íŒŒì•…
# ê²°ì¸¡ì¹˜: X
# ë²”ì£¼í˜•: ['branch', 'city', 'customer_type', 'gender', 'product_line', 'payment_method', 'time_of_day', 'day_name']
# target: total
# print(train.info())
# print(test.info())
# print(train)
# print(train.shape, test.shape) # (700, 10) (300, 9)
# print(train.select_dtypes('object').columns)

# 2. ë°ì´í„° ì „ì²˜ë¦¬ (4)
# (1) X_full
X = train.drop(['total'], axis=1)
y = train['total']
X_full = pd.concat([X, test], axis=0)

# (2) ê²°ì¸¡ì¹˜: X

# (3) ìˆ˜ì¹˜í˜• ë³€ìˆ˜ ìŠ¤ì¼€ì¼ë§: X

# (4) ë²”ì£¼í˜• ë³€ìˆ˜ ì¸ì½”ë”©
X_full = pd.get_dummies(X_full)
# print(X_full)
# print(X_full.shape) # (1000, 30)

# 3. ë°ì´í„° ë¶„ë¦¬ (X_train, X_test / train_test_split)
X_train = X_full[:train.shape[0]]
X_test = X_full[train.shape[0]:]
# print(X_train.shape, X_test.shape) # (700, 30) (300, 30)

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2)
# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape) # (560, 30) (140, 30) (560,) (140,)

# 4. ëª¨ë¸ í•™ìŠµ
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
RF.fit(X_train, y_train)
y_val_pred_rf = RF.predict(X_val)
# print(y_val_pred_rf)
# print(y_val_pred_rf.shape) # (140,)

from sklearn.ensemble import HistGradientBoostingRegressor
HGB = HistGradientBoostingRegressor()
HGB.fit(X_train, y_train)
y_val_pred_hgb = HGB.predict(X_val)
# print(y_val_pred_hgb)
# print(y_val_pred_hgb.shape) # (140,)

# 5. ëª¨ë¸ í‰ê°€
from sklearn.metrics import root_mean_squared_error
score_rf = root_mean_squared_error(y_val, y_val_pred_rf)
print('score_rf', score_rf)

score_hgb = root_mean_squared_error(y_val, y_val_pred_hgb)
print('score_hgb', score_hgb)

# 6. ê²°ê³¼ ì €ì¥
y_pred = RF.predict(X_test)
# print(y_pred)
# print(y_pred.shape) # (300,)

result = pd.DataFrame({'pred': y_pred})
result.to_csv('result_72.csv', index=False)

check = pd.read_csv('result_72.csv')
print(check)
# print(check.shape) # (300, 1)
```

## 8. ì„±ëŠ¥ ê°œì„ 

```python
# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/7_2/mart_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/7_2/mart_test.csv")

# target ë°ì´í„°
target = train.pop('total')

# ë ˆì´ë¸” ì¸ì½”ë”©
from sklearn.preprocessing import LabelEncoder
cols = train.select_dtypes(include='O').columns

for col in cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])

# ê²€ì¦ë°ì´í„° ë¶„ë¦¬
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

# ì„ í˜•íšŒê·€
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_tr, y_tr)
pred = lr.predict(X_val)
print(rmse(y_val, pred))

# ìµœì¢… ì œì¶œ íŒŒì¼
pred = lr.predict(test)
result = pd.DataFrame({
    'pred':pred
})
result.to_csv("result.csv", index=False)

# ë² ì´ìŠ¤ë¼ì¸(ì›í•« ì¸ì½”ë”©): 374016.3661648522
# ë ˆì´ë¸” ì¸ì½”ë”©: 371070.1344734549
```

## ë¬¸ì œ1
ì£¼ì–´ì§„ ì¡°ê°œ ë°ì´í„° 300ê°œ ì¤‘ ì•ì—ì„œë¶€í„° 210ê°œëŠ” train ë°ì´í„°ë¡œ ë§Œë“¤ê³ , ë‚˜ë¨¸ì§€ 90ê°œëŠ” testë°ì´í„°ë¡œ ë§Œë“ ë‹¤. ëª¨ë¸ì„ í•™ìŠµ(ì í•©)í•  ë•ŒëŠ” trainë°ì´í„°ë¥¼ ì‚¬ìš©í•˜ê³ , ì˜ˆì¸¡í•  ë•ŒëŠ” testë°ì´í„°ë¥¼ ì‚¬ìš©í•œë‹¤. ëª¨ë¸ì€ ë¡œì§€ìŠ¤í‹± íšŒê·€ë¥¼ ì¨ì„œ ì„±ë³„(gender)ì„ ì˜ˆì¸¡í•˜ë˜, íŒ¨ë„í‹°ëŠ” ë¶€ê³¼í•˜ì§€ ì•ŠëŠ”ë‹¤.

### ë¬¸ì œ1-1. weightë¥¼ ë…ë¦½ë³€ìˆ˜ë¡œ genderë¥¼ ì¢…ì†ë³€ìˆ˜ë¡œ ì‚¬ìš©í•˜ì—¬ ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨í˜•ì„ ë§Œë“¤ê³ , weight ë³€ìˆ˜ê°€ í•œ ë‹¨ìœ„ ì¦ê°€í•  ë•Œ ìˆ˜ì»·ì¼ ì˜¤ì¦ˆë¹„ ê°’ì€? (ë°˜ì˜¬ë¦¼í•˜ì—¬ ì†Œìˆ˜ ë„·ì§¸ìë¦¬ê¹Œì§€ ê³„ì‚°)

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/7_3/clam.csv")

# ë°ì´í„°ì…‹ ë¶„í• 
train = df.iloc[:210]
test = df.iloc[210:]
print(train.shape, test.shape)
```

```python
from statsmodels.formula.api import logit
import numpy as np

# 1) ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨ë¸ ìƒì„± ë° í•™ìŠµ
model = logit("gender ~ weight", data=train).fit()
print(model.summary())

# 2) ì˜¤ì¦ˆë¹„ ê³„ì‚°
print(round(np.exp(0.0047),4))
print(round(np.exp(model.params['weight']),4))
```

```python
# [ì°¸ê³ ] <ì‹¬í™”í•™ìŠµ> ì•”ì»·ì˜ ì˜¤ì¦ˆë¹„
female_odds_ratio = round(np.exp(-model.params['weight']), 4)
print(female_odds_ratio)
```

### ë¬¸ì œ1-2. genderë¥¼ ì¢…ì†ë³€ìˆ˜ë¡œ í•˜ê³  ë‚˜ë¨¸ì§€ ë³€ìˆ˜ë“¤(age, length, diameter, height, weight)ì„ ë…ë¦½ë³€ìˆ˜ë¡œ ì‚¬ìš©í•˜ëŠ” ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨ë¸ì„ ì í•©ì‹œí‚¨ í›„, ì”ì°¨ ì´íƒˆë„(Residual Deviance)ë¥¼ ê³„ì‚°í•˜ì‹œì˜¤. (ë°˜ì˜¬ë¦¼í•˜ì—¬ ì†Œìˆ˜ ë‘˜ì§¸ìë¦¬ê¹Œì§€ ê³„ì‚°)

```python
# ë°©ë²•1
from statsmodels.formula.api import logit

# 1) ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨ë¸ ìƒì„± ë° í•™ìŠµ
model = logit("gender ~ age + length + diameter + height + weight", data=train).fit()
print(model.summary())

# 2) ì”ì°¨ ì´íƒˆë„ ê³„ì‚°
print(round(-2 * model.llf,2))
```

```python
# ë°©ë²•2
from statsmodels.formula.api import glm
import statsmodels.api as sm

# 1) glm ëª¨ë¸ ì í•© (ë¡œì§€ìŠ¤í‹± íšŒê·€ë¥¼ ìœ„í•´ ì´í•­ ë¶„í¬ ì‚¬ìš©)
formula = "gender ~ age + length + diameter + height + weight"
model = glm(formula, data=train, family=sm.families.Binomial()).fit()

# 2) ì”ì°¨ì´íƒˆë„ ê³„ì‚°
print(model.summary())
print(round(model.deviance,2))
```

### ë¬¸ì œ1-3. ë…ë¦½ë³€ìˆ˜ weight ë§Œ ì‚¬ìš©í•´ í•™ìŠµí•œ ë¡œì§€ìŠ¤í‹± íšŒê·€ëª¨ë¸ì—ì„œ testë°ì´í„°ì˜ genderë¥¼ ì˜ˆì¸¡í•˜ê³ , error rate(ì˜¤ë¥˜ìœ¨)ë¥¼ êµ¬í•˜ì‹œì˜¤. (ë°˜ì˜¬ë¦¼í•˜ì—¬ ì†Œìˆ˜ ì…‹ì§¸ìë¦¬ê¹Œì§€ ê³„ì‚°)

```python
from sklearn.metrics import accuracy_score
from statsmodels.formula.api import logit

# ë°ì´í„°ì…‹ ë¶„í• 
train = df.iloc[:210]
test = df.iloc[210:]

# 1) í•™ìŠµ, testë°ì´í„°ë¥¼ ì‚¬ìš©í•´ ì˜ˆì¸¡ (0.5 ë¯¸ë§Œ: 0, 0.5 ì´ìƒ 1)
model = logit("gender ~ weight", data=train).fit()
target = test.pop("gender")
pred = model.predict(test) > 0.5

# 2) ì‹¤ì œ ê°’ê³¼ ì˜ˆì¸¡ ê°’ì„ ì‚¬ìš©í•˜ì—¬ ì •í™•ë„ ê³„ì‚°
acc = accuracy_score(target, pred)

# 3) ì˜¤ë¥˜ìœ¨ ê³„ì‚°
print(round(1-acc,3))
```

```python
# ë‹¤ë¥¸ í’€ì´ (ì‚¬ì´í‚·ëŸ°)
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

train = df.iloc[:210]
test = df.iloc[210:]

# 1) í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„° ë¶„ë¦¬
X_train = train[['weight']]
y_train = train['gender']
X_test = test[['weight']]
y_test = test['gender']

# 2) ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨ë¸ í•™ìŠµ
model = LogisticRegression()
model.fit(X_train, y_train)

# 3) í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¥¼ ì‚¬ìš©í•´ ì˜ˆì¸¡
pred = model.predict(X_test)

# 4) ì‹¤ì œ ê°’ê³¼ ì˜ˆì¸¡ê°’ì„ ì‚¬ìš©í•˜ì—¬ ì •í™•ë„ ê³„ì‚°
acc = accuracy_score(y_test, pred)

# 5) ì˜¤ë¥˜ìœ¨ ê³„ì‚°
print(round(1 - acc, 3))
```

### ë¬¸ì œ2-2. CPU ì»¬ëŸ¼ì´ 100ë¯¸ë§Œì¸ ê²ƒë§Œ ì°¾ì•„ ERPë¥¼ ì¢…ì† ë³€ìˆ˜ë¡œ, ë‚˜ë¨¸ì§€ ë³€ìˆ˜ë“¤ì„ ë…ë¦½ ë³€ìˆ˜ë¡œ ì„¤ì •í•´ ì„ í˜•íšŒê·€ ëª¨ë¸ì„ ë§Œë“¤ê³  ì í•©í•œ ê²°ì •ê³„ìˆ˜ë¥¼ êµ¬í•˜ì‹œì˜¤. (ë°˜ì˜¬ë¦¼í•˜ì—¬ ì†Œìˆ˜ ì…‹ì§¸ìë¦¬ê¹Œì§€ ê³„ì‚°)

```python
from statsmodels.formula.api import ols

# 1) CPUê°€ 100 ë¯¸ë§Œì¸ ë°ì´í„° í•„í„°ë§
cond = df['CPU'] < 100
df = df[cond]

# 2) ì„ í˜•íšŒê·€ ëª¨ë¸ ìƒì„±: ERPë¥¼ ì¢…ì† ë³€ìˆ˜ë¡œ, ë‚˜ë¨¸ì§€ ë³€ìˆ˜ë“¤ì„ ë…ë¦½ ë³€ìˆ˜ë¡œ ì„¤ì •
model = ols('ERP ~ Feature1 + Feature2 + Feature3 + CPU', data=df).fit()

# 3) ëª¨ë¸ ìš”ì•½ ì •ë³´ ì¶œë ¥
print(model.summary())
print(model.rsquared)
print(model.rsquared_adj)
```

```python
# pvalue
# 0.457
print(model.pvalues)
```
